## 목소리 추출하기

```python
# 목소리 추출하기

import os
# 현재 위치 저장
now_path = os.getcwd()

# 폴더 이동
print(os.getcwd())
try:
    os.chdir('./music_download/wav_files/')
except:
    pass

# 파일 목록 가져오기
files = os.listdir(os.getcwd())

for file in tqdm(files):
    # 음성 파일만 작동
    if '.wav' in file:
        # 분리하기
        spl = f'spleeter separate -p spleeter:2stems -o output {file}'
        os.system(spl)
        # spleeter separate -p spleeter:2stems -o output 4AM4AM.wav
        try:
            # 분리한 음성 파일, 멜로디 옮기기
            # 음성 파일 이름 바꾸기
            os.rename(f'./output/{file[:-4]}/vocals.wav', f'./output/{file[:-4]}/{file}')
            os.rename(f'./output/{file[:-4]}/accompaniment.wav', f'./output/{file[:-4]}/{file[:-4]}_accompaniment.wav')
            
            # 폴더 삭제
            # shutil.move(f'./output/{file[:-4]}/{file}', f'../../change_music/diff-svc/')
            # shutil.move(f'./output/{file[:-4]}/', f'../spleeter_vocals/')
            os.system(f'del ./output/{file[:-4]}')
            
            shutil.move(f'./output/{file[:-4]}/{file}', f'../spleeter_vocals')
            shutil.move(f'./output/{file[:-4]}/{file[:-4]}_accompaniment.wav', f'../bgm_files/')
            # 사용한 파일은 다른 곳으로 옮기기
            shutil.move(f'./{file}', f'../use/{file}')
        except FileExistsError:
            print('파일이 이미 존재합니다.')
        
os.chdir('../../')
```
